# 选择版本空间

这里采用的找版本空间的方法是：保留((与正例一致的假设)&(与反例不一致的假设))  
这里采用二进制按位与的方法进行条件判断,这也是为什么 * 必须要用11表示的原因  
因为11与00，01，10按位与结果均为11，这对应通配符 * 包含所有属性的涵义

```python
def get_versionspace(data):
    # 将object类型转为hex类型
    data['hex'] = [0x3f,0x3d,0x3e,0x37,0x3b,0x1f,0x2f,0x35,0x36,0x39,0x3a,0x1d,0x1e,0x2d,0x2e,
                               0x17,0x1b,0x27,0x2b,0x15,0x16,0x19,0x1a,0x25,0x26,0x29,0x2a]
    # 题中所说的编号1(正例)和4(反例)的两个样例
    sample = data['hex'][[19,26]] 
    # 与正例一致的假设
    condition1 = (data['hex']|sample[19]) == data['hex']
    # 与反例不一致的假设
    condition2 = (data['hex']|sample[26]) != data['hex']
    return data[(conditon1)&(condition2)]
```

```python
version_space = get_versionspace(hypothesis_space)
version_space
```
